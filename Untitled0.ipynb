```python
# ============================================
# AI-Based Electricity Theft Detection System
# ============================================

# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# --------------------------------------------
# Step 1: Generate Electricity Consumption Data
# --------------------------------------------

np.random.seed(42)

consumer_id = list(range(1001, 1101))

monthly_consumption = np.random.normal(250, 40, 100)
peak_usage = np.random.normal(80, 15, 100)
billing_amount = np.random.normal(1500, 300, 100)

# Inject abnormal values (simulate theft cases)
monthly_consumption[95:100] = [700, 750, 800, 820, 900]
billing_amount[95:100] = [400, 420, 450, 470, 500]

# Create DataFrame
data = pd.DataFrame({
    "Consumer_ID": consumer_id,
    "Monthly_Consumption_kWh": monthly_consumption,
    "Peak_Usage_kWh": peak_usage,
    "Billing_Amount": billing_amount
})

print("Sample Dataset:")
print(data.head())

# --------------------------------------------
# Step 2: Select Features for Model
# --------------------------------------------

features = data[[
    "Monthly_Consumption_kWh",
    "Peak_Usage_kWh",
    "Billing_Amount"
]]

# --------------------------------------------
# Step 3: Train Isolation Forest Model
# --------------------------------------------

model = IsolationForest(
    n_estimators=100,
    contamination=0.05,
    random_state=42
)

model.fit(features)

# --------------------------------------------
# Step 4: Predict Electricity Theft
# --------------------------------------------

data["Anomaly"] = model.predict(features)

data["Status"] = data["Anomaly"].apply(
    lambda x: "Suspected Theft" if x == -1 else "Normal"
)

print("\nDetection Results (Last 10 Consumers):")
print(data[["Consumer_ID", "Status"]].tail(10))

# --------------------------------------------
# Step 5: Visualization of Results
# --------------------------------------------

plt.figure(figsize=(8,5))
plt.scatter(
    data["Monthly_Consumption_kWh"],
    data["Billing_Amount"],
    c=data["Anomaly"]
)

plt.xlabel("Monthly Consumption (kWh)")
plt.ylabel("Billing Amount")
plt.title("Electricity Theft Detection using Isolation Forest")
plt.show()
```
